```python
import numpy as np
import random
import scipy.stats as sps
```

```python
a = np.random.sample((10,1000))
b = np.random.sample((1000,10))
```

```python
np.ndim(a), np.ndim(b)
```

```python
np.shape(a), np.shape(b)
# Функция shape вывела размеры матриц
```

```python
def mult(a, b):
    if len(a[0]) == len(b):
        c = [[0 for x in range(len(a))] for y in range(len(b[0]))]
        for x in range(len(a)):
            for y in range(len(b[0])):
                c[x][y] = sum(u * v for u,v in zip(a[x, :],b[:, y]))
        return c
    else:
        print("Произведение указанных матриц недопустимо")
```

```python
def np_mult(a, b):
    if np.shape(a)[1] == np.shape(b)[0]:
        c = np.dot(a, b)
        return c
    else:
        print("Произведение указанных матриц недопустимо")
```

```python
%%time
 # засечем время работы функции без NumPy 
mult(a, b)
```

```python
%%time
# засечем время работы функции с NumPy
np_mult(a, b)
```

```python
# функция, решающая задачу с помощью NumPy
def sec_av(A):
    s = np.zeros_like(A)
    for k in range(len(A)):
        s[k] = np.average(A[: k + 1])
    return s
```

```python
# функция без NumPy
def stupid_sec_av(A):
    s = [0 for i in range(len(A))]
    for i in range(len(A)):
        s[i] = sum(u for u in A[: i + 1])/(i + 1)
    return s
```

```python
# зададим некоторую последовательность и проверим ее на ваших функциях. 
# Первая функция должна работать ~ в 50 раз быстрее
A = sps.uniform.rvs(size=10 ** 3) 

%time S1 = sec_av(A)
%time S2 = stupid_sec_av(A)
#проверим корректность:
np.abs(S1 - S2).sum()
```

```python
# функция, решающая задачу с помощью NumPy
def transformation(X, a=1):
    Y = np.empty_like(X)
    d = 1 + (len(X) - 1) % 2
    for k in range(len(X)):
        if k % 2 == 1:
            Y[k] = a
        else:
            Y[len(X) - k - d] = X[k] ** 3
    Y = np.concatenate([X, Y])
    Y = np.flip(Y, axis = 0)
    return Y
```

```python
# функция, решающая задачу  без NumPy
def stupid_transformation(X, a=1):
    Y = [None] * len(X) * 2
    d = 1 - len(X) % 2
    e = len(Y) - 1
    for k in range(len(X)):
        Y[e - k]=X[k]
        if k % 2 == 1:
            Y[k - d] = a
        else:
            Y[k + d] = X[k] ** 3
    return Y
```

```python
X = sps.uniform.rvs(size=10 ** 7) 
%time S1 = transformation(X, 2)
%time S2 = stupid_transformation(X, 2)
# проверим корректность:
np.abs(S1 - S2).sum()
```

```python
#Задача 4
from matplotlib import pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

def f(x, y):
    '''Функция двух переменных'''
    return np.sin(x) * np.cos(y)

def draw_f(grid_x, grid_y):
    '''Функция отрисовки функции f(x, y)'''
    fig = plt.figure(figsize=(10, 8))
    ax = Axes3D(fig)
    ax.plot_surface(grid_x, grid_y, f(grid_x, grid_y), cmap='inferno')
    plt.show()
```

```python
grid_x, grid_y = np.meshgrid(np.arange(-10,10), np.arange(-10,10))

draw_f(grid_x, grid_y)
```
