```python
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import numpy as np
```

```python
import scipy.stats as st
import seaborn as sns
y = train['SalePrice']
plt.figure(1); plt.title('Normal')
sns.distplot(y, kde=False, fit=st.norm)
plt.figure(2); plt.title('Log Normal')
sns.distplot(y, kde=False, fit=st.lognorm)
```

С какими типами признаков вы столкнулись?

### Часть 3
Какие признаки выглядят бесполезными? Попробуйте выбросить некоторые фичи, опираясь на ваш личный жизненный опыт или алгоритмы `feature selection`. Подробнее про них можно прочитать здесь: http://scikit-learn.org/stable/modules/feature_selection.html

Разделим признаки на категориальные и количественные

```python
quantitative = [f for f in train.columns if train.dtypes[f] != 'object']
quantitative.remove('SalePrice')
quantitative.remove('Id')
qualitative = [f for f in train.columns if train.dtypes[f] == 'object']
```

Пропуски среди категориальныхх признаков

```python
#количество NaN в разрезе признаков
missing = train[qualitative].isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar()
```

Пропуски среди количественных признаков

```python
missing = train[quantitative].isnull().sum()
missing = missing[missing > 0]
missing.sort_values(inplace=True)
missing.plot.bar()
```

Заполняем пропуски

```python
x_train_qual = train[qualitative].copy()
x_train_quant = train[quantitative].copy()
x_train_quant.fillna(0, inplace=True)
x_train_qual.fillna('NA', inplace=True)

x_test_qual = test[qualitative].copy()
x_test_quant = test[quantitative].copy()
x_test_quant.fillna(0, inplace=True)
x_test_qual.fillna('NA', inplace=True)
```

С помощью дисперсионного анализа можно посмотреть, какие признаки оказывают большее влияние на цену, 
а какие почти не оказывают влияния

```python
def anova(frame):
    anv = pd.DataFrame()
    anv['feature'] = qualitative
    pvals = []
    for c in qualitative:
        samples = []
        for cls in frame[c].unique():
            s = frame[frame[c] == cls]['SalePrice'].values
            samples.append(s)
        pval = st.f_oneway(*samples)[1]
        pvals.append(pval)
    anv['pval'] = pvals
    return anv.sort_values('pval')

sample = pd.concat([x_train_qual, y_train_log], axis=1)
a = anova(sample)
a['disparity'] = np.log(1./a['pval'].values)
sns.barplot(data=a, x='feature', y='disparity')
x=plt.xticks(rotation=90)
```

Удалим наименее влияющие на цену признаки

```python
x_train_qual = x_train_qual.drop(["Utilities", "LandSlope", "Street"], axis=1)
```

Созание дамми по категориальным признакам

```python
x_train_qual_dummy = pd.get_dummies(x_train_qual)
x_test_qual_dummy = pd.get_dummies(x_test_qual)
```

```python
x_train_fin = pd.concat([x_train_quant, x_train_qual_dummy], axis=1)
x_test_fin = pd.concat([x_test_quant, x_test_qual_dummy], axis=1)
list(x_train_fin)
```

Для отбора переменных будем использовать ExtraTreesClassifier

```python
from sklearn.ensemble import ExtraTreesClassifier
from sklearn.feature_selection import SelectFromModel
clf = ExtraTreesClassifier()
clf = clf.fit(x_train_fin, y_train.values.ravel())
model = SelectFromModel(clf, prefit=True)
model = SelectFromModel(clf, prefit=True)
x_names_fin = x_train_fin.columns[model.get_support()]
x_train_fin[x_names_fin].shape
```

Построим графики, отражающие зависимость цены от наиболее значимых с моей точки зрения переменных.

```python
def pairplot(x, y, **kwargs):
    ax = plt.gca()
    ts = pd.DataFrame({'time': x, 'val': y})
    ts = ts.groupby('time').mean()
    ts.plot(ax=ax)
    plt.xticks(rotation=90)

significant_features = ['LotArea', 'YearBuilt', 'Neighborhood']
f = pd.melt(train, id_vars=['SalePrice'], value_vars=significant_features)
g = sns.FacetGrid(f, col="variable",  col_wrap=2, sharex=False, sharey=False, size=5)
g = g.map(pairplot, "value", "SalePrice")
```

С помощью  *cross_val_score* можно оценивать качество построенной модели.
Также у каждого регрессора есть множество *параметров*. Подробно про них можно прочитать в документации http://scikit-learn.org

Пример запуска модели можно найти в **Seminars/Seminar2**.

```python
# функция для расчета оценочной метрики
from sklearn.model_selection import KFold, cross_val_score
n_folds = 6
def rmse_val(model, train):
    kf = KFold(n_folds, shuffle=True, random_state=21).get_n_splits(train.values)
    rmse= np.sqrt(-cross_val_score(model, train.values, np.log1p(labels), scoring="neg_mean_squared_error", cv = kf))
    return(rmse)
```

Рассчитаем метрики для линейной регрессии, решающего леса и градиентного бустинга

```python
from sklearn.linear_model import LinearRegression
from sklearn.ensemble import RandomForestRegressor
from sklearn.ensemble import   GradientBoostingRegressor

regr = LinearRegression()
regr.fit(x_train_fin[x_names_fin], y_train_log)
score = rmse_val(regr, x_train_fin[x_names_fin])
print(score.mean(), score.std())

rforest = RandomForestRegressor(max_depth=13, random_state=21, n_estimators=18)
rforest.fit(x_train_fin[x_names_fin], y_train_log.values.ravel())
score = rmse_val(rforest, x_train_fin[x_names_fin])
print(score.mean(), score.std())

GBoost = GradientBoostingRegressor(learning_rate=0.06, max_depth=6, min_samples_leaf=6, loss='huber', random_state = 21)
GBoost.fit(x_train_fin[x_names_fin], y_train_log.values.ravel())
score = rmse_val(GBoost, x_train_fin[x_names_fin])
print(score.mean(), score.std())
```

```python
#функция для формирования файла для отправки в соревнование
preds = np.expm1(GBoost.predict(x_test_fin[x_names_fin])).ravel()

submission = pd.DataFrame()
submission["Id"]= test['Id'].ravel()
submission["SalePrice"] = preds
submission.to_csv("submission.csv", index=False)
```
